```python
## âœ… Python Code

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(file_path)

# Convert Sale_Date to datetime
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

# 1. Total Sales

total_sales = df["Sales_Amount"].sum()
print("Total Sales:", total_sales)


# 2. Sales by Region

sales_by_region = df.groupby("Region")["Sales_Amount"].sum()
print("\nSales by Region:")
print(sales_by_region)

# 3. Sales by Product Category

sales_by_category = df.groupby("Product_Category")["Sales_Amount"].sum()
print("\nSales by Product Category:")
print(sales_by_category)

# 4. Monthly Sales Trend

df["Month"] = df["Sale_Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales_Amount"].sum()

# Plot monthly sales
plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.show()

### ðŸ”¹ Top Sales Representative

top_rep = df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values(ascending=False)
print("\nTop Sales Representatives:")
print(top_rep)


### ðŸ”¹ Total Profit

df["Profit"] = (df["Unit_Price"] - df["Unit_Cost"]) * df["Quantity_Sold"]
total_profit = df["Profit"].sum()
print("\nTotal Profit:", total_profit)
```
